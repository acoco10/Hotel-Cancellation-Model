```python
X = df.drop('is_canceled', axis =1)
y = df.is_canceled
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=99)
```

```python
from imblearn.under_sampling import TomekLinks
tl = TomekLinks()
X_resampled, y_resampled = tl.fit_sample(X_train, y_train)
X_test_resampled, y_test_resampled = tl.fit_sample(X_test, y_test)
```

```python
scaler = MinMaxScaler()
scaled = scaler.fit_transform(X_resampled)
scaled_test = scaler.transform(X_test_resampled)
```

```python
lr = LogisticRegression (penalty = 'l1', solver='saga', max_iter = 1000)
lr.fit(scaled,y_resampled)
test_pred = lr.predict(scaled_test)
```

```python
metrics.precision_score(y_test_resampled, test_pred)
```

```python
lr.coef_
```

```python
print(metrics.confusion_matrix(y_test, test_pred))
```

```python
confusion = metrics.confusion_matrix(y_test, test_pred)
TP = confusion[1][1]
TN = confusion[0][0]
FP = confusion[0][1]
FN = confusion[1][0]
```

```python
print ('True Positives:', TP)
print ('True Negatives:', TN)
print ('False Positives:', FP)
print ('False Negatives:', FN)
```
